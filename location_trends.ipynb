```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_excel("UMERA BOOST 2.0 & 3.0.xlsx", sheet_name="UMRA 3.0")

# Clean column names (remove extra spaces)
df.rename(columns=lambda x: x.strip(), inplace=True)

# Standardizing location column
location_column = "Location"  # Ensure this is the correct column name in your dataset
df[location_column] = df[location_column].astype(str).str.strip().str.lower()  # Convert to lowercase & strip spaces

# Replace common variations with standardized names
location_corrections = {
    "lagos, nigeria": "lagos",
    "ibadan, oyo state": "ibadan",
    "ogbomosho": "ogbomoso",  # Fix spelling variation
    "lagos mainland": "lagos",
    "lagos island": "lagos",
    "ibadan - ui": "ibadan",
    "ibadan.": "ibadan",  # Remove periods at the end
    "lagos.": "lagos",
    "abuja, fct": "abuja"
}

df[location_column] = df[location_column].replace(location_corrections)

# Count the number of participants per location
location_distribution = df[location_column].value_counts()

# Get the top 10 locations
top_10_locations = location_distribution.head(10)

# Visualization: Top 10 Locations of Registrants
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_locations.index, y=top_10_locations.values, hue=top_10_locations.index, palette="magma", legend=False)
plt.xlabel("Location")
plt.ylabel("Number of Participants")
plt.title("Top 10 Locations of Registrants")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()


# Display numerical distribution
print("Location Distribution:\n")
print(location_distribution)

# Percentage breakdown
location_percentage = (location_distribution / location_distribution.sum()) * 100
print("\nPercentage Breakdown of Locations:\n")
print(location_percentage.round(2).astype(str) + '%')
```

Location Trends Analysis Report
Introduction
Understanding the geographical distribution of participants is critical for optimizing the outreach and impact of UMRAs Corporate Social Responsibility (CSR) initiatives. This analysis examines the locations from which registrants for the UMRA 3.0 program originate, highlighting key trends that can inform future engagement and expansion strategies.

Key Findings
1. Dominance of Ibadan and Lagos
The highest number of registrants come from Ibadan, accounting for the two largest bars in the visualization. Combined, Ibadan contributes 192 registrants (106 + 86), making it the most dominant location.
Lagos follows closely with 115 registrants (75 + 40), further solidifying its position as a major participant hub.
These two cities together account for a significant portion of the total registrants, indicating that UMRAs influence is strongest in these areas.
2. Emerging Locations with Moderate Participation
Ogbomoso and Abuja appear as secondary hubs, contributing 42 and 35 participants, respectively.
The presence of multiple registrations from different locations within the same city (e.g., Lagos appearing twice in the ranking) suggests that participants may be registering under varied sub-regions, leading to slight data fragmentation.
3. Long-Tail Distribution of Other Locations
Several other locations contribute smaller numbers of registrants, including Abeokuta, Osogbo, and other scattered towns.
The distribution suggests that while UMRAs program has a stronghold in specific urban areas, its penetration into other regions remains relatively limited.
Insights and Strategic Recommendations
1. Strengthen Engagement in Ibadan and Lagos
Since these two cities account for a significant majority of registrants, efforts should be made to consolidate and expand the existing audience base.
Strategies such as localized events, partnerships with educational institutions, and targeted social media campaigns can be leveraged to maintain dominance in these areas.
2. Address Data Standardization Issues
The presence of duplicate city names (e.g., Lagos appearing twice) suggests the need for better data cleaning before analysis.
Implementing a pre-processing step to standardize location names in future data collection efforts will improve data accuracy.
3. Expand Outreach in Secondary Cities
Cities like Ogbomoso and Abuja, though smaller in registrant numbers, show potential for growth.
Targeted campaigns and partnerships in these areas could help bridge the participation gap between them and the larger hubs.
4. Broaden Regional Outreach
Since participation from other cities is relatively sparse, a focused effort to increase awareness and accessibility in these regions is necessary.
Hosting satellite events, deploying community ambassadors, or leveraging digital campaigns may help reach new participants.
Conclusion
This location trends analysis highlights that Ibadan and Lagos are the strongest participant hubs for UMRAs CSR initiative, with Ogbomoso and Abuja emerging as secondary locations. However, opportunities exist to expand outreach into smaller towns and to improve data standardization for better insights. By leveraging these findings, UMRA can optimize its engagement strategies, improve regional representation, and ensure the program's impact is more widespread.
